# Off-Platform Project: Viral Tweets

In this project, we are going to use the K-Nearest Neighbor algorithm to predict whether a tweet will go viral. 

Let's explore these options by looking at the data we have available to us. We've imported the dataset and printed the following information:

* The total number of tweets in the dataset.
* The columns, or features, of the dataset.
* The text of the first tweet in the dataset.


````python
import pandas as pd
pd.set_option('max_columns', None)
from pprintpp import pprint as pp
import warnings
warnings.filterwarnings('ignore')
import numpy as np
from sklearn.preprocessing import scale, StandardScaler, MinMaxScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn import model_selection
from xgboost import XGBClassifier
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV
from sklearn.model_selection import RepeatedStratifiedKFold, RepeatedKFold



import matplotlib.pyplot as plt
````

````python
all_tweets = pd.read_json("random_tweets.json", lines=True)

print(len(all_tweets))
pp(all_tweets.columns)
pp(all_tweets.loc[0]['text'])
print('\n')
pp(all_tweets.loc[0, 'user'])
print('\n')
pp(all_tweets.loc[:, 'entities'])
display(all_tweets.head())
````

# Defining Viral Tweets


So how do we define a viral tweet? A good place to start is to look at the number of retweets the tweet has. This can be found using the feature `"retweet_count"`. Let's say we wanted to create a column called `is_viral` that is a `1` if the tweet had more than `5` retweets and `0` otherwise. We could do that like this:

```py
all_tweets['is_viral'] = np.where(all_tweets['retweet_count'] > 5, 1, 0)
```

Instead of using `5` as the benchmark for a viral tweet, let's use the median number of retweets. You can find that by calling the `median()` function on `all_tweets["retweet_count"]`. 



````python

all_tweets['is_viral'] = np.where(
    all_tweets['retweet_count'] > 10000, 1, 0)

#np.median(all_tweets['retweet_count']
pp(all_tweets.is_viral.value_counts())
````

# Making Features

Now that we've created a label for every tweet in our dataset, we can begin thinking about which features might determine whether a tweet is viral. We can create new columns in our dataset to represent these features. For example, let's say we think the length of a tweet might be a valuable feature. The following line creates a new column containing the length of the tweet.


We can create a new column called `followers_count` that contains the number of followers of each user. We do the same for `friends_count`.

We can also add other features that we feel might play a crucial role.


````python
all_tweets['tweet_length'] = all_tweets.apply(lambda tweet: len(tweet['text']), axis=1)
all_tweets['num_hashtags'] = all_tweets.apply(lambda tweet: tweet['text'].count('#'), axis=1)
all_tweets['num_links'] = all_tweets.apply(lambda tweet: tweet['text'].count('http'), axis=1)
all_tweets['followers_count'] = all_tweets.apply(lambda tweet: tweet['user']['followers_count'], axis=1)
all_tweets['friends_count'] = all_tweets.apply(lambda tweet : tweet['user']['friends_count'], axis = 1)
all_tweets['is_verified'] = all_tweets.apply(lambda tweet : 1 if(tweet['user']['verified']) else 0, axis = 1)
print('\n\n\n')
display(all_tweets.head())
````

# Normalizing The Data


````python
class Model:
    def __init__(self, data, labels):
        self.data = data
        self.labels = labels

    def split_data(self, test_size):
        self.x_train, self.x_test, self.y_train, self.y_test = train_test_split(
            self.data, self.labels, test_size=test_size, random_state=1)

    def knnClassify_accuracy(self, n_neighbors):
        classifier = KNeighborsClassifier(n_neighbors=n_neighbors)
        classifier.fit(self.x_train, self.y_train)
        return classifier.score(self.x_test, self.y_test)
    
    def lrClassify_accuracy(self):
        classifier = LogisticRegression()
        classifier.fit(self.x_train, self.y_train)
        return classifier.score(self.x_test, self.y_test)
    
    def svmClassify_accuracy(self):
        classifier = SVC()
        classifier.fit(self.x_train, self.y_train)
        return classifier.score(self.x_test, self.y_test)
    
    def dTreeClassify_accuracy(self):
        classifier = DecisionTreeClassifier()
        classifier.fit(self.x_train, self.y_train)
        return classifier.score(self.x_test, self.y_test)
    
    def RForestClassify_accuracy(self):
        classifier = RandomForestClassifier()
        classifier.fit(self.x_train, self.y_train)
        return classifier.score(self.x_test, self.y_test)
    
    def XGBClassify_accuracy(self):
        classifier = XGBClassifier()
        classifier.fit(self.x_train, self.y_train)
        return classifier.score(self.x_test, self.y_test)
````

````python
labels = all_tweets[['is_viral']]
data = all_tweets[[
    'tweet_length', 'num_hashtags', 'num_links', 'followers_count',
    'friends_count', 'is_verified'
]]


standard_scaled_classifier = StandardScaler()
standard_scaled_data = standard_scaled_classifier.fit_transform(data)
````

# Creating the Training Set and Test Set



````python

data = Model(standard_scaled_data, labels)
data.split_data(0.2)
````

# Using the Classifier



````python

print(f'\n Logistic Regression Score: {data.lrClassify_accuracy()}')
print(f'\n Support Vector Machines Score: {data.svmClassify_accuracy()}')
print(f'\n KNN Score: {data.knnClassify_accuracy(5)}')
print(f'\n Decision Tree Score: {data.dTreeClassify_accuracy()}')
print(f'\n Random Forest Score : {data.RForestClassify_accuracy()}')
print(f'\n XGBClassifier Score : {data.XGBClassify_accuracy()}')
````

# Compare the models

As can be observed from above, the accuracy keeps increasing as we increase the threshold for a tweet to be considered as viral. All the required models can also be compared using `model_selection.cross_val_score`

````python
seed = 7

models = []
models.append(('Logistic Regression :', LogisticRegression()))
models.append(('KNN :', KNeighborsClassifier()))
models.append(('Decision Tree :', DecisionTreeClassifier()))
models.append(('Random Forest :', RandomForestClassifier()))
models.append(('SVM :', SVC()))
models.append(('XGBoost :', XGBClassifier(verbosity = 0)))

results = []
names = []
scoring = 'accuracy'
for name, model in models:
    kfold = model_selection.KFold(n_splits=10, random_state=seed)
    cv_results = model_selection.cross_val_score(model, standard_scaled_data, labels, cv=kfold, scoring=scoring)
    results.append(cv_results)
    names.append(name)
    msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())
    print(msg)
# boxplot algorithm comparison
fig = plt.figure()
fig.suptitle('Algorithm Comparison')
ax = fig.add_subplot(111)
plt.boxplot(results)
ax.set_xticklabels(names)
plt.xticks(rotation = 90)
plt.show()

````

# Hyperparameter Tuning

We have now found out that XGBoost classifier has produced the best results. Let's find out the best parameters for XGBoost that give us the best accuracy.

````python
train_data, test_data, train_labels, test_labels = train_test_split(
    standard_scaled_data, labels, test_size=0.2, random_state=1)

classifier = XGBClassifier(verbosity = 0)
param_grid = {
    'max_depth': [6, 10, 15, 20],
    'learning_rate': [0.001, 0.01, 0.1, 0.2, 0, 3],
    'subsample': [0.5, 0.6, 0.7, 0.8, 0.9, 1.0],
    'colsample_bytree': [0.4, 0.5, 0.6, 0.7, 0.8, 0.9, 1.0],
    'colsample_bylevel': [0.4, 0.5, 0.6, 0.7, 0.8, 0.9, 1.0],
    'min_child_weight': [0.5, 1.0, 3.0, 5.0, 7.0, 10.0],
    'gamma': [0, 0.25, 0.5, 1.0],
    'reg_lambda': [0.1, 1.0, 5.0, 10.0, 50.0, 100.0],
    'n_estimators': [100]
}
# fit_params = {'eval_metric': 'mlogloss',
#               'early_stopping_rounds': 10}

cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)

search = RandomizedSearchCV(classifier,
                            param_grid,
                            n_iter=50,
                            n_jobs=-1,
                            scoring='accuracy',
                            random_state=1)

search.fit(train_data, train_labels)

print('Best Score: %s' % search.best_score_)
print('Best Hyperparameters: %s' % search.best_params_)
````
