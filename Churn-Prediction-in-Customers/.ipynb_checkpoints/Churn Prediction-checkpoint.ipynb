# Churn Prediction

Retention rate is one of the most important metric to assess the success of a particular business on a monthly basis. Churn prediction is one of the techniques that helps us to quantify the retention rate and do the needful for retaining the customers which are predicted to be churning.

<br>
As compared to the previous parts, we will be changing our dataset as the dataset considered in the previous parts does not have the data whether the customers were retained or not.

```python
def plot_vs_churn(col_name, var_type):

        df_plot = df.groupby([col_name])['Churn'].mean().reset_index()

        if var_type == 'categorical':
            sns.barplot(x = df_plot[col_name],y = df_plot['Churn'])
            if col_name == 'PaymentMethod':
                plt.xticks(rotation = 90)
            plt.show()

        elif var_type == 'numerical':
            sns.scatterplot(data = df_plot, x = col_name, y = 'Churn')
            plt.show()
```

```python
cat_columns = ['gender', 'InternetService', 'Contract', 'PaymentMethod', 'PaperlessBilling',
    'StreamingMovies', 'DeviceProtection', 'PhoneService'
]

for col_name in cat_columns:
    plot_vs_churn(col_name, 'categorical')
```

```python
plot_vs_churn('tenure', 'numerical')
```

```python
sse={}
df_cluster = df[['tenure', 'MonthlyCharges', 'TotalCharges']]
df_cluster['TotalCharges'] = pd.to_numeric(df_cluster['TotalCharges'], downcast = 'float', errors = 'coerce')

df_cluster = df_cluster.dropna()
print(df_cluster['TotalCharges'].isna().value_counts())

df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], downcast = 'float', errors = 'coerce')

df = df.dropna()
```

```python
scaler = MinMaxScaler()
df_cluster = scaler.fit_transform(df_cluster)
display(df_cluster)
```

```python
for k in range(1, 10):
    kmeans = KMeans(n_clusters=k, max_iter=1000).fit(df_cluster)
    sse[k] = kmeans.inertia_ 
plt.figure()
plt.plot(list(sse.keys()), list(sse.values()))
plt.xlabel("Number of cluster")
plt.show()
```

```python

kmeans = KMeans(n_clusters=3)
kmeans.fit(df[['tenure']])
df['tenureCluster'] = kmeans.predict(df[['tenure']])
kmeans.fit(df[['MonthlyCharges']])
df['MonthlyChargesCluster'] = kmeans.predict(df[['MonthlyCharges']])
kmeans.fit(df[['TotalCharges']])
df['TotalChargesCluster'] = kmeans.predict(df[['TotalCharges']])

```

```python
display(df)
```

Fucntion to sort the clusters in an ascending order

```python
def order_cluster(cluster_field_name, target_field_name, df, ascending):
    df_cluster_order = df.groupby([cluster_field_name
                                   ])[target_field_name].mean().reset_index()
    df_cluster_order = df_cluster_order.sort_values(
        by=target_field_name, ascending=ascending).reset_index(drop=True)
    df_cluster_order['index'] = df_cluster_order.index
    
    df_final = pd.merge(df, df_cluster_order[[cluster_field_name, 'index']], on = cluster_field_name)
    df_final = df_final.drop([cluster_field_name], axis = 1)
    df_final = df_final.rename(columns={"index": cluster_field_name})
    return df_final
    
```

```python
cols = ['tenure', 'MonthlyCharges', 'TotalCharges']
for col in cols:
    df = order_cluster( col+'Cluster', col, df, True)

display(df)
```

Now that we have sorted the clusters in an ascending order of their means, let's double check if it was right.


```python
for col in cols:
    pp(df.groupby([col+'Cluster'])[col].describe())
```

<br>

```python
cols_rename = ['tenureCluster', 'MonthlyChargesCluster', 'TotalChargesCluster']
for col in cols_rename:
    df[col] = df[col].map({0: 'Low', 1: 'Med', 2: 'High'})
```

#### Encoding categorical values

```python
df.info()
```

```python
#import Label Encoder
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
dummy_columns = [] #array for multiple value columns

for column in df.columns:
    if df[column].dtype == object and column != 'customerID':
        if df[column].nunique() == 2:
            #apply Label Encoder for binary ones
            df[column] = le.fit_transform(df[column]) 
        else:
            dummy_columns.append(column)

#apply get dummies for selected columns
df = pd.get_dummies(data = df,columns = dummy_columns)
```

```python
display(df)
```

## Logistic Regression Model

Predicting whether a customer is churned or retained is a binary classification problem hence we will be using the logistic regression algorithm to further delve into the details.
<br>
First let's create a logistic regression model and then get the summary of the model as shown below:

```python
all_columns = []
for column in df.columns:
    column = column.replace(" ", "_").replace("(", "_").replace(")", "_").replace("-", "_")
    all_columns.append(column)

df.columns = all_columns
```

```python
glm_columns = 'gender'

for column in df.columns:
    if column not in ['Churn','customerID','gender']:
        glm_columns = glm_columns + ' + ' + column
```

```python
pp(glm_columns)
```

```python

import statsmodels.api as sm
import statsmodels.formula.api as smf
 

glm_model = smf.glm(formula='Churn ~ {}'.format(glm_columns), data=df, family=sm.families.Binomial())
res = glm_model.fit()
print(res.summary())
```

### Observations

From the above observations, we need to asnwer 2 questions regarding our churn rate:
 - Which characteristics makes the customer churn or retain?
     - If the P > |z| value in the summary is less than our p-value (0.05), it significantly affects the churn.
 - What are the most critical ones that we need to focus on?
     - This question can be reframed as "Which feature brings the best ROI if i increase/decrease it by one unit. This can be found in the coef_ paramter of the summary. Exponential coef gives us the expected change in Churn Rate if we change it by one unit.

```python
np.exp(res.params)
```

As an example, one unit change in Monthly Charge means ~3.4% improvement in the odds for churning if we keep everything else constant. From the table above, we can quickly identify which features are more important.
Now, everything is ready for building our classification model.

## XGBoost Classifier for prediction

```python
from sklearn.model_selection import train_test_split
import xgboost as xgb
#create feature set and labels
X = df.drop(['Churn','customerID'],axis=1)
y = df.Churn
#train and test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.05, random_state=56)
#building the model & printing the score

xgb_model = xgb.XGBClassifier(max_depth=5, learning_rate=0.08, objective= 'binary:logistic',n_jobs=-1).fit(X_train, y_train)
print('Accuracy of XGB classifier on training set: {:.2f}'
       .format(xgb_model.score(X_train, y_train)))
print('Accuracy of XGB classifier on test set: {:.2f}'
       .format(xgb_model.score(X_test[X_train.columns], y_test)))
```

```python
from sklearn.metrics import classification_report
y_pred = xgb_model.predict(X_test)

print(classification_report(y_test, y_pred))
```

From the above report we observe that the precision seems to be a bit low which can be improved by using more features, better feature engineering and hyperparameter tuning.

We need to get a basic idea of all the features and their importances that were used by our model:

```python
from xgboost import plot_importance
fig, ax = plt.subplots(figsize=(10,8))
plot_importance(xgb_model, ax=ax)
```

We can see that our model assigned more importance to TotalCharges and MonthlyCharges compared to others.
<br>Finally, the best way to use this model is assigning Churn Probability for each customer, create segments, and build strategies on top of that.

```python
df['proba'] = xgb_model.predict_proba(df[X_train.columns])[:,1]
df_customer_churn = df[['customerID', 'proba']]
display(df_customer_churn)
```
